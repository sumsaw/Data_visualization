Reading the downloaded and unzipped files programatically

## Multivariate Exploration


 Lets group tine into 4 zones Night(0-6) , Morning (6-12),afternoon(12-6) and evening(6-12)

```python
bike_share_2018['start_hour']=(pd.to_datetime(bike_share_2018['start_time_dt'],format= '%H:%M:%S' ).dt.hour)
```

```python
# Slotting time into various zones of the day as described above 
bin_edges = [0,6,12,18,24]
bin_labels=['Night','Morning','Afternoon','Evening']
```

```python
bike_share_2018['day_zone']=pd.cut(bike_share_2018['start_hour'],bin_edges,labels=bin_labels)
```

```python
bike_share_2018.sample(5)
```

```python
plt.figure(figsize=(11.69,8.27));
g = sns.FacetGrid(data = bike_share_2018, col = 'day_zone',height= 4,col_order=['Morning','Afternoon','Evening','Night']);
g.map(sns.barplot, 'member_gender', 'duration_sec');
#plt.ylabel('Average Duration( secs)');
```

We see that women ride higher in all the four day zones with maximum difference in ride times in afternoon and least in the night<br> 

Morning ( 6-12) , Afternon(12-18), Evening(18-0) ,Night(0-6), Bracker numbers are time in 24hrs format<br>

Understanding why women ride are longer duration on average can help us push male ride duration up as well
